```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = './network_trace/'
networks = []
for i in range(20):
    network = pd.read_csv(file_path+str(i),delimiter = ' ',names=['time','throughput'])
    networks.append(network)
```

```python
fig = plt.figure(dpi=100,figsize=(12,80),constrained_layout=True)
for i in range(len(networks)):
    network = networks[i]
    ax = fig.add_subplot(20,1,i+1)
    ax.set_ylim(0,6)
    ax.plot(network['time'],network['throughput'])
    print(network['throughput'].describe())
plt.show()
```
